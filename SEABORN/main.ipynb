```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('iris')
```

#  TEMEL SEVİYE

## 1. GRAFİK TÜRLERİ

#### 1.1  Dağılım

```python
# Dağılım grafiği
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

#### 1.2 Histogram

```python
# Histogram
sns.histplot(df['sepal_length'], bins=30, kde=True)
plt.show()
```

#### 1.3 Kutu

```python
# Kutu grafiği
sns.boxplot(data=df, x='species', y='sepal_length')
plt.show()
```

#### 1.4 Isı

```python
# Korelasyon matrisini hesapla
corr = df.corr()

# Isı haritası oluştur
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

#### 1.5 Çizgi

```python
# Çizgi grafiği
sns.lineplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

## 2.Grafik Özelleştirme


#### 2.1 Başlık ve Eksen Etiketleri

```python
# Grafik ile başlık ve eksen etiketleri ekleyin
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

#### 2.2 Renk Paletleri

```python
# Renk paletini değiştirme
sns.set_palette('husl')
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

#### 2.3 Grafik Stilleri

```python
# Grafik stilini değiştirme
sns.set_style('darkgrid')
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

## 3 Çoklu Grafikler


```python
# FacetGrid kullanarak çoklu grafikler
g = sns.FacetGrid(df, col='species')
g.map(sns.scatterplot, 'sepal_length', 'sepal_width')
plt.show()
```

# İLERİ SEVİYE


# 1. Kullanıcı Tanımlı Renk Paletleri
Renk paletlerini özelleştirebilir ve kendi renk paletinizi oluşturabilirsiniz. Bu, grafiklerinizin daha estetik ve tutarlı görünmesini sağlar.

```python
# Kendi renk paletinizi oluşturun
custom_palette = sns.color_palette(['#FF5733', '#33FF57', '#3357FF'])

# Renk paletini kullanın
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette=custom_palette)
plt.show()
```

# 2. Grafiklerin İç İçe Geçmesi (Subplots)
Seaborn ve Matplotlib'i birlikte kullanarak birden fazla grafiği aynı figür içinde gösterebilirsiniz.

```python
# Alt grafikler (subplots) oluşturma
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# İlk grafik
sns.histplot(df['sepal_length'], bins=30, kde=True, ax=axes[0], color='blue')
axes[0].set_title('Sepal Length Distribution')

# İkinci grafik
sns.boxplot(data=df, x='species', y='sepal_length', ax=axes[1])
axes[1].set_title('Sepal Length by Species')

plt.tight_layout()
plt.show()
```

# 3. Kapsayıcı Veri Kümesi (PairPlot)
PairPlot, bir veri kümesindeki tüm çiftler arasındaki ilişkileri görselleştirir ve ayrıca her bir değişkenin histogramını gösterir.

```python
# PairPlot kullanarak tüm çiftler arasındaki ilişkileri gösterin
sns.pairplot(df, hue='species')
plt.show()
```

# 4. Kategorik Veri Üzerinde Yoğunluk Eğrisi (Violin Plot)
Violin plot, veri dağılımını ve yoğunluğunu hem sayısal hem de kategorik veri üzerinde gösterir.

```python
# Violin plot kullanarak veri dağılımını ve yoğunluğunu gösterme
sns.violinplot(data=df, x='species', y='sepal_length', inner='quartile')
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

# 5. Kümeleme (Clustering) ve Renk Kodlama
Isı haritasında, verilerinizi kümeler halinde renklendirerek daha anlamlı görselleştirmeler oluşturabilirsiniz.

```python
# Kümeleme (clustering) ile ısı haritası oluşturma
corr = df.corr()

sns.clustermap(corr, annot=True, cmap='coolwarm', fmt='.2f', center=0, linewidths=0.5, cbar_kws={"shrink": .8})
plt.title('Clustered Heatmap')
plt.show()
```

# 6. Koordinat Sistemleri ve Eğilim Çizgileri
Çizgi grafiklerinde veya dağılım grafikleri üzerinde eğilim çizgileri ekleyebilirsiniz.

```python
# Dağılım grafiğine eğilim çizgisi ekleme
sns.regplot(data=df, x='sepal_length', y='sepal_width', scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title('Scatter Plot with Regression Line')
plt.show()
```

# 7. Yüzey Grafikleri (3D Plotlar)
Seaborn doğrudan 3D grafikler sunmaz, ancak Matplotlib ile birlikte kullanarak 3D grafikler oluşturabilirsiniz.

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 3D yüzey grafiği
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Veri
x = df['sepal_length']
y = df['sepal_width']
z = np.random.rand(len(df)) * 10  # Rastgele z verisi

# 3D yüzey grafiği
ax.scatter(x, y, z, c='b', marker='o')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Random Z')
plt.show()
```

# 8. Gelişmiş Estetik ve Temalar
Grafiklerin estetik görünümünü özelleştirebilir ve temaları değiştirebilirsiniz.

```python
# Estetik ve tema ayarları
sns.set_style('whitegrid')
sns.set_palette('pastel')

# Basit bir dağılım grafiği
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Styled Scatter Plot')
plt.show()
```
