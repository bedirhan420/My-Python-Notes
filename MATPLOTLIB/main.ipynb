# 1. Alt Grafikleri Kullanma (Subplots)
Matplotlib'de birden fazla grafiği tek bir figürde göstermek için subplots kullanabilirsiniz. Bu, verileri karşılaştırmak için çok kullanışlıdır.

```python
import matplotlib.pyplot as plt
import numpy as np

# Veri hazırlığı
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Alt grafikler oluşturma
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(x, y)
axs[0, 0].set_title('Grafik 1')

axs[0, 1].plot(x, -y)
axs[0, 1].set_title('Grafik 2')

axs[1, 0].plot(x, y ** 2)
axs[1, 0].set_title('Grafik 3')

axs[1, 1].plot(x, np.cos(x))
axs[1, 1].set_title('Grafik 4')

plt.tight_layout()
plt.show()
```

# 2. Stil Özelleştirmeleri
Matplotlib'de grafiklerinizi daha estetik hale getirmek için stilleri kullanabilirsiniz.

```python
plt.style.use('ggplot')  # ggplot stilini kullanma

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Stil Örneği')
plt.show()
```

# 3. Grafiklerde Annotasyonlar
Grafiklerinize önemli noktaları vurgulamak için anotasyonlar ekleyebilirsiniz.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Anotasyon Örneği')

# Anotasyon ekleme
plt.annotate('Maksimum', xy=(1.57, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

# 4. Üç Boyutlu Grafikler
Matplotlib ile üç boyutlu grafikler oluşturabilirsiniz.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```

# 5. Kolaylaştırıcı İpuçları
Matplotlib Kısayol Stilleri: Kendi stil dosyalarınızı oluşturup kullanarak grafiklerinizin tutarlılığını sağlayabilirsiniz.
Büyük Veri Setleri ile Çalışma: Büyük veri setleriyle çalışırken, agg backend'ini kullanmak performansı artırabilir.
Jupyter Notebook ile Entegrasyon: %matplotlib inline komutunu kullanarak grafiklerin doğrudan not defterinizde görüntülenmesini sağlayabilirsiniz.
Hata Çubukları: Verilerinize hata çubukları ekleyerek daha bilgilendirici grafikler oluşturabilirsiniz.

```python
# Hata çubukları ekleme
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = np.random.rand(10) * 0.1

plt.errorbar(x, y, yerr=yerr, fmt='-o')
plt.title('Hata Çubukları Örneği')
plt.show()
```

# 6. Grafiklere İkinci Eksen Ekleme
Aynı grafikte iki farklı eksen kullanarak farklı ölçekteki verileri görselleştirebilirsiniz.

```python
fig, ax1 = plt.subplots()

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x / 10)

ax1.plot(x, y1, 'b-')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)', color='b')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'r-')
ax2.set_ylabel('exp(x / 10)', color='r')

plt.title('İkinci Eksen Örneği')
plt.show()
```
